## 1.) Equation of Circle with center at $(a,b)$ and radius $r$
 \begin{equation}(x-a)^2+(y-b)^2=r^2 \end{equation}
## 2.) Parametric form
\begin{cases} x=a+r\cos{(\theta)}\\  y=b+r\sin{(\theta)}\end{cases}

<div>

</div>

## Conditions to be followed:-
<ol>
  <li>Number of petals = Number of Circles</li>
  <li>All circles of equal radius.</li>
  <li>All circles uniformly distributed across $2\pi$ radian.</li>
  <li>For $n$ number of petals, the angular shift between the center of each consecutive circle will be
    \begin{equation} \frac{2\pi}{n}\end{equation}</li>
</ol>

<hr style="border:2px solid black">

```python
#Import the libraries
import numpy as np
from matplotlib import pyplot as plt, patches
```

```python
#Graph a circle of radius=3 using parametric equations
r=3
t = np.linspace(0, 2 * np.pi)

#Parametric equation
x, y = r * np.cos(t), r * np.sin(t)
fig, ax = plt.subplots()
ax.plot(x, y, "k")
ax.set(aspect=1)
plt.grid()
plt.show()
```

```python
#Graph 4 petals using circle of radius=3 using parametric equations
r=3
t = np.linspace(0, 2 * np.pi)
theta=2*np.pi/4 # for 4 petals the angular shift will be 2*pi/4

fig, ax = plt.subplots(figsize=(8, 8))
for i in range(0,4,1):
    x, y = r*np.cos(i*theta) + r * np.cos(t), r*np.sin(i*theta) + r * np.sin(t)
    ax.plot(x, y,'k')

plt.grid()
plt.show()
```
