# Get the equation and graph of the circle

```python
#Import the libraries
import numpy as np
import matplotlib.pyplot as plt 
from sympy import Circle, latex
```

```python
#The code to get the user input for the center of the circle
while True:
    try:
        user_input = input('Enter comma-separated integers in the form a,b: ')
        center = tuple(int(item) for item in user_input.split(','))
        if len(center)==2:
            pass
        else:
            continue
    except:
        print("Looks like you did not enter an integer!")
        continue
    else:
        print("Yep those are the coordinates!")
        break

a,b=center[0],center[1]
print("The coordinates of the center of the circle is ({},{})".format(a,b))
```

```python
#The code to get the user input for the radius of the circle
while True:
    try:
        r = float(input('Enter the radius of the circle: '))
    except:
        print("Looks like you did not enter an integer!")
        continue
    else:
        print("Yep that's the value!")
        break
print("The radius of the circle is {}".format(r))
```

```python
#Retrive the equation of the circle
c = Circle((a, b), r)
equation = c.equation()
latex_code = latex(equation)

plt.text(0.5, 0.5, f'${latex_code}=0$', size=30)
plt.axis('off')
plt.show()
```

```python
#Graph a circle using parametric equations
t = np.linspace(0, 2 * np.pi,100)

#Parametric equation
x, y = a+r * np.cos(t), b+r * np.sin(t)
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y, "k")
ax.set(aspect=1)
plt.grid()
plt.show()
```
