```python
#Import the libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Graph 24 petals using circle of radius=3 using parametric equations
r=3
t = np.linspace(0, 2 * np.pi)
theta=2*np.pi/24 # for 24 petals the angular shift will be 2*pi/24

fig, ax = plt.subplots(figsize=(12, 12))
for i in range(0,24,1):
    x, y = r*np.cos(i*theta) + r * np.cos(t), r*np.sin(i*theta) + r * np.sin(t)
    ax.plot(x, y,'k')

plt.grid()
plt.show()
```
